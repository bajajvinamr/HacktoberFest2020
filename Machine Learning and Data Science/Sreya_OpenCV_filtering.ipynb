```
#importing the required libraries

import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```
# importing a sample Image for demonstrating filter implementaion
# imread is a function we use for reading image 
img = cv2.imread('/content/Sample_1_2.tif')
plt.imshow(img)
# here I have used plt.imshow() so that I can get good subplots
# Instead of this we can use cv2.imshow()
```

## FILTERING AN IMAGE



* We use filters to smoothen the image,remove noises,blur image , find edges of objects in image


#### cv2.filter2D() 


This function can use for anytype of filter 

```
# specify filter :
# here for example I'm taking a average filter ,will explain the working of average filter in next steps
filter = np.ones((3,3),np.float32)/9
```

```
img_output1 = cv2.filter2D(img,-1,filter)
plt.imshow(img_output1,cmap='gray')
```

### Averaging filter :
* It is a low pass filter used for blurring ( smoothening image ) 


```
img_averagefilter = cv2.blur(img,(5,5))
```

```
# using subplots to compare the input image and output image
plt.figure(figsize=(14,12))
plt.subplot(221)
plt.imshow(img)
plt.title('input_image')
plt.subplot(222)
plt.imshow(img_averagefilter)
plt.title('after_average_filter')
```

Here you can observe that the details are blurred(smoothned) from input_image 



### Median filter
* Low pass filter
* This is used for smoothening image by preserving small deatils 


```
# here 5 is filter size that is 5*5
img_medianfilter = cv2.medianBlur(img,5)
```

```
plt.figure(figsize=(14,12))
plt.subplot(221)
plt.imshow(img)
plt.title('input_image')
plt.subplot(222)
plt.imshow(img_medianfilter)
plt.title('after_medianfilter')
```

You can observe here that even though it blurred the image still it has preserved some small details ( small dots in this case)

### Laplacian filtering


*   High pass filter used for edge detection 



```
# importing a different image for easy understanding of edge detection
img_1 = cv2.imread("/content/SanFrancisco.jpg",0)
#here 0 specified in above step will take grayscale image for better understanding of filtering
```

```
img_laplace = cv2.Laplacian(img_1,-1,ksize=3)
# 5 is size of filter that is 5*5
# here there are some preselected filters which we calling through function by mentioning size of that
```

```
plt.figure(figsize=(14,12))
plt.subplot(221)
plt.imshow(img_1,cmap='gray')
plt.title('input_image')
plt.subplot(222)
plt.imshow(img_laplace,cmap='gray')
plt.title('after_laplace')
```

Here you can observe the edges detected in this 2D image using single laplacian filter

### SobelGradient filtering
* high pass filtering
* It is similar to laplace filtering but here we can specify in which dimension you want to detect edges

```
img_sobel_y = cv2.Sobel(img_1,-1,0,1,3)
plt.imshow(img_sobel_y,cmap='gray')
```

```
# HORIZONTAL 
img_sobel_x = cv2.Sobel(img_1,-1,1,0,3)
plt.imshow(img_sobel_x,cmap='gray')
```

Here you can observe 1st figure shows you x-dimensional edges whereas second one shows y-dimensional edges
